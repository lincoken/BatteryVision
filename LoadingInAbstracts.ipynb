# These are my attempts to load in the abstracts

```python
from gensim.models import Word2Vec
import numpy as np
import html2text 
import codecs
from bs4 import BeautifulSoup as bs
from html.parser import HTMLParser
```

## First Try: html2text

```python
filename = "Web of Science [v.5.31] - WOS Export Transfer Service.html"
f=codecs.open(filename, 'r')
abstracts = f.read()
raw_text = html2text.html2text(abstracts)
#print(raw_text)
print(type(raw_text))
```

```python
# printing to text file to load in Word2Vec file
text_file = open("Abstacts.txt", "w")

text_file.write(raw_text)

text_file.close()
```

## Second Try: Beautiful Soup

```python
soup = bs(abstracts, 'html.parser')
soup.title
print(soup.get_text().strip())
```
